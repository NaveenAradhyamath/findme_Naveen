## Objective : To predict the future revenues, expenses and  gross profits of a company using Monte Carlo simulation

Importing the required modules 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Assigning to the variables required for the analysis

Assuming the following values of past performance of the company

mean - 320

standard deviation - 20

Iteration - 1000 values

revenue value in million dollars

```python
revenue_mean = 320
revenue_std = 20
iterations = 1000
```

I have used numpy's 'random' method to generate different series of values 

```python
rev = np.random.normal(revenue_mean,revenue_std,iterations)
print(rev)
```

Plot showing the different revenue values generated

```python
plt.figure(figsize = (15,6))
plt.plot(rev)
plt.show()
```

Cost on goods sold is the expenditure spent on goods, materials and manufature  of the company

I have assumed 60% of company's COGS value
and standard deviation to be 10%

since we have already generated 1000 values we can just multiply with revenue by using the numpy random method.

```python
COGS = - (rev * np.random.normal(0.6,0.1))
plt.figure(figsize = (15,6))
plt.plot(COGS)
plt.show()
```

```python
COGS.mean()
```

```python
COGS.std()
```

To compute the gross profit value 

since i have used negative value for the cogs so i have added to the revenue. which operates a substraction from revenue.

```python
Gross_Profit = rev + COGS
Gross_Profit
```

```python
max(Gross_Profit)
```

```python
min(Gross_Profit)
```

mean value of gross profit using mean method

```python
Gross_Profit.mean()
```

```python
Gross_Profit.std()
```

Plot showing the series of distribution of predicted Gross_Profit values

```python
plt.figure(figsize = (15,6))
plt.hist(Gross_Profit, bins = 20, color = 'r')
plt.title('Gross Profit Distribution')
plt.ylabel('series of observation')
plt.xlabel('Gross profit values')
plt.show()
```

### Conclusion: We can see that the maximum distribution of the values fall between the range 140-150 and hence we have found out the mean that is 144 and we can conclude that the predicted value of gross profit is 144 million dollars
