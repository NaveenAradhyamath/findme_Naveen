```python
df1 = df[(df['baths'] != 4) & (df['rooms'] != 10) | (df['rooms'] != 4) 
         & (df['nbh'] != 3) & (df['age'] <= 100)] # removing the outliers
df1.head()
```

```python
df1.describe()
```

```python
sns.distplot(df1['age'], kde = True,  bins =3)
```

```python
my_tab4 = pd.crosstab(index=df1["age"],  
                     columns="count")                  
my_tab4.sort_values(by = 'count', ascending = False).head(20)
```

```python
df1['age'] = np.where(df1['age'] >= 3, 3,df1['age']) #handling with outliers
```

```python
df1.describe()
```

```python
df1['cbd'] = np.log(df1['cbd'])
```

```python
df_new = df1[['year','age','nbh', 'cbd', 'lintst',
       'rooms', 'baths', 'ldist', 'wind', 'lprice',
       'y81', 'larea', 'lland', 'y81ldist', 'lintstsq', 'nearinc', 'y81nrinc', 'lrprice ']]
```

```python
df_new.head()
```

```python
df_new.describe()
```

```python
f, ax = plt.subplots(figsize=(15,8))
sns.heatmap(df_new.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

```python
plt.plot(df_new.year,df_new.lprice, linestyle = 'solid', label = 'Price', marker = 'o', color = 'r')

plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')

plt.title('year vs price')
plt.legend(fontsize = 16, loc = 0)
plt.show()
print()
```

```python
fig = plt.figure(figsize=(20,5))
df_new[['year','y81nrinc']].groupby(['year']).sum().plot(figsize=(15,5))
```
